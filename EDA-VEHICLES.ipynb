```python
# Checking for Missing Values
df.isnull().sum()
```

```python
# convert zip code column into integer
df['Zip Code']=pd.to_numeric(df['Zip Code'],errors='coerce')
```

```python
df['Zip Code'].isnull().sum()
```

```python
# filling zip code column with the mode value
df['Zip Code'] = df['Zip Code'].fillna(df['Zip Code'].value_counts().index[0])
```

```python
df['Zip Code'].isnull().sum()
```

## Let's summarize our data by row, column, features, unique, and missing values.

```python
print ("Rows     : " ,df.shape[0])
print ("Columns  : " ,df.shape[1])
print ("\nFeatures : \n" ,df.columns.tolist())
print ("\nUnique values :  \n",df.nunique())
print ("\nMissing values :  ", df.isnull().sum().values.sum())
```

```python
# Convert the Date Feature Column to a Datetime Format
df['Date'] =  pd.to_datetime(df['Date'],
                              format='%m/%d/%Y')
```

```python
df['Date'].info()
```

```python
# Now we will parse Date into three columns that is year, month, and day.
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
```
